```python
import pandas as pd
df = pd.read_csv('german.data.csv', index_col=None, header=None)
df.columns = ['checking_account', 'duration_months', 'credit_history', 'purpose', 'credit_amount', 'savings_account', 'employment_since', 'installment_rate', 'personal_status_sex', 'other_debtors', 'residence_since', 'property', 'age', 'other_installment_plans', 'housing', 'number_credits', 'job', 'people_liable', 'telephone', 'foreign_worker', 'credit_risk']
df.head(10)
```

```python
#Using german.data, involving preprocessing, splitting and knn
# Splitting data from target
target = df.iloc[:,-1]
data = df.iloc[:,:-1]
data.head(10)


from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report


#Separating data from target
dfnum = pd.read_csv('german.data.csv', index_col=None, header=None)
X = dfnum.iloc[:,:-1]
y = dfnum.iloc[:,-1]
X_encoded = pd.get_dummies(X, drop_first=True)
X_encoded.columns = X_encoded.columns.astype(str)


#Splitting data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.25, random_state = 42, stratify = y
)


#Scaling each training data with mean 0 and standard deviation 1
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)

#Scaling test data, prevents data leakage by using transform
#Uses the same mean and standard deviation as the training data to scale the test data
X_test_scaled = scaler.transform(X_test)


#KNN
knn = KNeighborsClassifier(n_neighbors=10)
knn.fit(X_train_scaled, y_train)
y_pred = knn.predict(X_test_scaled)
print(y_pred)
print('accuracy', accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
#Using german.data-nummeric, only involving splitting and knn
#Separating data from target
dfnum = pd.read_csv('german.data-numeric.csv', index_col=None, header=None)
X = dfnum.iloc[:,:-1]
y = dfnum.iloc[:,-1]


#Splitting data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state = 42, stratify = y
)


#Scaling each training data with mean 0 and standard deviation 1
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)

#Scaling test data, prevents data leakage by using transform
#Uses the same mean and standard deviation as the training data to scale the test data
X_test_scaled = scaler.transform(X_test)


#KNN
knn = KNeighborsClassifier(n_neighbors=10)
knn.fit(X_train_scaled, y_train)
y_pred = knn.predict(X_test_scaled)
print(y_pred)
print('accuracy', accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
